```python
file_path1 = r"T:\Study\TAMIL\ds1.csv"  
file_path2 = r"T:\Study\TAMIL\ds2.csv"


d1 = pd.read_excel(file_path1, engine='openpyxl')
d2 = pd.read_excel(file_path2, engine='openpyxl')

print(d1.head())
```

```python
print(d2.head())
```

```python
d1['property_type'] = "sale"
d2["property_type"] = "rent"
print(d1)
```

```python
d1=d1.drop("Per_Sqft",axis=1)
d1.info()
```

```python
import numpy as np
def split_floor(value):
    try:
        value = value.strip()
        parts = value.lower().split("out of")
        current = parts[0].strip()
        total = parts[1].strip()
        if current == "ground":
            current = 0
        else:
            current = int(current)
        total = int(total)
        return current, total

    except:
        return np.nan, np.nan
d2["current_floor"], d2["total_floors"] = zip(*d2["Floor"].apply(split_floor))
d2.drop("Floor", axis=1, inplace=True)
```

```python
d2.info()
```

```python
missing_cols = set(d1.columns) - set(d2.columns)

for col in missing_cols:
    d2[col] = None
```

```python
missing_in_d1 = set(d2.columns) - set(d1.columns)

for col in missing_in_d1:
    d1[col] = None

```

```python
d1.info()
```

```python
d2.info()
```

```python
d1 = d1[sorted(d1.columns)]
d2 = d2[sorted(d2.columns)]
```

```python
final = pd.concat([d1, d2], ignore_index=True)
final.info()
```

```python
object_cols = final.select_dtypes(include=['object']).columns
final[object_cols] = final[object_cols].fillna("unknown")

numeric_cols = final.select_dtypes(include=['float64','int64']).columns
final[numeric_cols] = final[numeric_cols].fillna(0)
```

```python
final.info()
```

```python
print(final.to_string())
```
