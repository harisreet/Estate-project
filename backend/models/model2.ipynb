```python
import pandas as pd
```

```python
file_path2 = r"T:\Study\TAMIL\ds2.csv"
df = pd.read_excel(file_path2, engine='openpyxl')

print(df.head())
```

```python
df = df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
print(df.to_string)
```

```python
import numpy as np
def split_floor(value):
    try:
        value = value.strip()
        parts = value.lower().split("out of")
        current = parts[0].strip()
        total = parts[1].strip()
        if current == "ground":
            current = 0
        else:
            current = int(current)
        total = int(total)
        return current, total

    except:
        return np.nan, np.nan
df["current_floor"], df["total_floors"] = zip(*df["floor"].apply(split_floor))
df.drop("floor", axis=1, inplace=True)
df['current_floor'].fillna(df['current_floor'].median(), inplace=True)
df['total_floors'].fillna(df['total_floors'].median(), inplace=True)
```

```python
# sensible ranges — tweak to your domain
df = df[(df['price'] >= 2000) & (df['price'] <= 200000)]   # price monthly
df = df[(df['area']  >= 100)  & (df['area']  <= 5000)]     # sqft
df = df[(df['bhk']   >= 0)    & (df['bhk']   <= 10)]
```

```python
df['price_per_sqft'] = df['price'] / df['area']
df['floor_ratio'] = df['current_floor'] / df['total_floors']
df['floor_ratio'] = df['floor_ratio'].fillna(0)
df['is_high_floor'] = (df['current_floor'] > (df['total_floors'] * 0.5)).astype(int)
df['bathroom_per_bhk'] = df['bathroom'] / df['bhk'].replace(0, 1)
```

```python
cat_cols = [
    "area_type",
    "furnishing",
    "tenant_preferred",
]
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))  
    encoders[col] = le

joblib.dump(encoders, "encoders2.pkl")
```

```python
import numpy as np
y = np.log1p(df['price'])
```

```python
x=df.drop(columns=["price"],axis=1)
y=df["price"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
     n_estimators=200,       # smaller forest
    max_depth=15,          # limits complexity
    min_samples_split=10,
    min_samples_leaf=5,
    max_features='sqrt',
    random_state=42
)
model.fit(X_train, y_train)

```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
train_score = model.score(X_train, y_train)
print("Training R2 Score:", train_score)
```

```python
test_score = model.score(X_test, y_test)
print("Test R2 Score:", test_score)
```

```python
import pickle

with open("rent_model.pkl", "wb") as f:
    pickle.dump(model, f)
```
